```python
import pandas as pd
import numpy as np
import pyam
import matplotlib.pyplot as plt
```

```python
url = 'https://github.com/Tschikko/python-projects/blob/cbc73a386d1546c1a73141946ff11a455f031c18/iamc-1.5c-explorer_snapshot_1635725378.csv?raw=true'
csv = pd.read_csv(url)
csv.head()
```

```python
df = pyam.IamDataFrame(data=csv)
```

```python
df.filter(variable='Agricultural Production|Non-Energy*', region='World').plot(color='variable')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

df_world_co2 = df.filter(variable='Agricultural Production|Non-Energy*')

df_world_co2.plot(ax=ax[0])
df_world_co2_2 = df.filter(variable='Agricultural Production|Energy*').plot(ax=ax[1])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

df_world_co2 = df.filter(variable='Agricultural Demand|Crops|Energy')

df_world_co2.plot(ax=ax[0])
df_world_co2_2 = df.filter(variable='Agricultural Demand|Crops|Food').plot(ax=ax[1])
```

```python
df.rename(variable={'AR5 climate diagnostics|Temperature|Global Mean|MAGICC6|MED': 'Temperature'},
          inplace=True)
```

```python
df.filter(variable='Temperature').plot()
```

```python
df.set_meta(meta='uncategorized', name='warming-category')
```

```python
df.meta.head()
```

```python
df.categorize(
    'warming-category', '1.5C',
    criteria={'Temperature': {'up': 1.5, 'year': 2100}},
    color='xkcd:baby blue'
)
```

```python
df.categorize(
    'warming-category', '1.75C',
    criteria={'Temperature': {'up': 1.75, 'lo':1.5, 'year': 2100}},
    color='xkcd:green'
)
```

```python
df.categorize(
    'warming-category', '2C',
    criteria={'Temperature': {'up': 2, 'lo':1.75, 'year': 2100}},
    color='xkcd:goldenrod'
)
```

```python
df.categorize(
    'warming-category', 'over 2C',
    criteria={'Temperature': {'lo':2, 'year': 2100}},
    color='xkcd:crimson'
)
```
